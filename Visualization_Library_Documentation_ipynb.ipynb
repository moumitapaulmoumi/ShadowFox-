### ðŸ”· Matplotlib

- **Introduction:**  
  Matplotlib is one of the oldest and most widely used data visualization libraries in Python. It provides extensive functionalities for creating static, animated, and interactive plots in Python.

- **History & Background:**  
  Developed by John D. Hunter in 2003, Matplotlib was originally designed to mimic MATLAB's plotting capabilities and offer an open-source alternative.

- **Key Features:**
  - Supports a wide range of plots: line, bar, scatter, pie, histogram, contour, 3D plots, and more.
  - Highly customizable: full control over figure size, fonts, colors, ticks, legends, annotations, etc.
  - Can save plots in many formats: PNG, PDF, SVG, EPS, etc.
  - Integrates well with **NumPy**, **Pandas**, and **Jupyter Notebooks**.
  - Supports interactive plotting with **IPython** and GUI toolkits (like Tkinter, Qt).

- **Architecture:**
  - Matplotlib has two main interfaces:
    1. **Pyplot API** â€“ A MATLAB-style interface for quick plotting (`import matplotlib.pyplot as plt`)
    2. **Object-Oriented API** â€“ Offers more control and flexibility for advanced plots.

- **Use Cases:**
  - Scientific research and engineering
  - Exploratory Data Analysis (EDA)
  - Reports and dashboards
  - Academic publication-quality figures

- **Limitations:**
  - Default visuals are basic (but can be styled manually)
  - Steeper learning curve for advanced customization
  - Static plots (interactivity is limited without extra tools)

- **Example Use Cases:**
  - Time series analysis
  - Plotting mathematical functions
  - Comparing categories in grouped bar charts
  - Displaying frequency distributions via histograms

- **Common Plot Types in Matplotlib:**
  - Line Plot (`plt.plot()`)
  - Bar Chart (`plt.bar()`)
  - Scatter Plot (`plt.scatter()`)
  - Histogram (`plt.hist()`)
  - Pie Chart (`plt.pie()`)
  - Box Plot (`plt.boxplot()`)

- **Basic Example:**
  ```python
  import matplotlib.pyplot as plt

  x = [1, 2, 3, 4, 5]
  y = [2, 4, 6, 8, 10]

  plt.plot(x, y, color='green', marker='o', linestyle='--')
  plt.title('Line Plot Example')
  plt.xlabel('X Axis')
  plt.ylabel('Y Axis')
  plt.grid(True)
  plt.show()


### ðŸ“ˆ 2D Line Plot using Matplotlib

#### âœ… Description:
A line plot is one of the most basic and commonly used plots in data visualization. It displays information as a series of data points (called "markers") connected by straight lines.

#### ðŸŸ¢ When to Use:
- To visualize trends or changes over time.
- When comparing one or more series of data.
- For showing continuous data like temperature, stock prices, or sensor readings.

#### ðŸ“Œ Why Use in Matplotlib:
- Easy to create with `plt.plot()`
- Fully customizable (line styles, markers, colors, grids, titles)
- Supports multiple lines on the same graph


````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create the plot
plt.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2)

# Add title and labels
plt.title("Simple 2D Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Show grid and plot
plt.grid(True)
plt.show()
````

### ðŸ“Š Multi-Line Plot using Matplotlib

#### âœ… Description:
A multi-line plot displays multiple line graphs on the same axes. Each line represents a different dataset or category, making it easy to compare trends or patterns side by side.

#### ðŸŸ¢ When to Use:
- To compare two or more sets of time-series or continuous data.
- To analyze multiple trends in one graph (e.g., sales across multiple stores).
- For visual comparisons between different variables.

#### ðŸ“Œ Why Use in Matplotlib:
- Allows plotting multiple lines by simply calling `plt.plot()` multiple times.
- Fully customizable (colors, legends, markers, etc.)
- Useful for storytelling with comparative insights.


````python
import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
product_A = [100, 120, 140, 160, 180]
product_B = [90, 130, 150, 170, 160]

# Plotting multiple lines
plt.plot(months, product_A, label='Product A', color='blue', marker='o')
plt.plot(months, product_B, label='Product B', color='green', marker='s')

# Add labels, title, legend
plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````

### ðŸ§© Figure and Axes in Matplotlib (Object-Oriented API)
#### ðŸ”¹ Step 2: Python Code Cell (Basic Figure & Axes Line Plot)


#### âœ… Description:
Instead of using the simpler `pyplot` interface (`plt.plot()`), Matplotlib also provides an object-oriented approach where you create a `Figure` and one or more `Axes`. This is especially useful for:
- Complex layouts with multiple plots
- Fine-tuned control over positioning and customization

#### ðŸŸ¢ When to Use:
- When creating subplots or multiple plots in one figure
- When you need precise control over each plot area
- For dashboard-like visualizations or publication-quality plots

#### ðŸ“Œ Components:
- **Figure**: The entire canvas where plots live.
- **Axes**: An individual plot (a part of the figure) â€” like one graph inside the canvas.

#### ðŸŽ¯ Syntax Overview:
```python
fig, ax = plt.subplots()
ax.plot(x, y)


````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]

# Create Figure and Axes
fig, ax = plt.subplots()

# Plot on the Axes
ax.plot(x, y, color='purple', marker='^', linestyle='-', label='Line A')

# Customize the plot using Axes methods
ax.set_title("Line Plot using Figure and Axes")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.grid(True)
ax.legend()

# Display the plot
plt.show()
````

####  **ðŸ”¹Vertical Bar Chart**

````python

#### ðŸ”¹ Step 2: Python Code Cell (Basic Vertical Bar Chart Example)


import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 33]

# Create a vertical bar chart
plt.bar(categories, values, color='orange')

# Add title and labels
plt.title("Simple Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")

# Show the plot
plt.show()
````

````python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 33]

# Create a horizontal bar chart
plt.barh(categories, values, color='green')

# Add title and labels
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Category")

# Show the plot
plt.show()
````

####  **ðŸ”¹Grouped Bar Chart**
   - **Description**: Compares two or more sets of data side by side within each category.
   - **Use Case**: Used to compare multiple data sets (e.g., sales in different regions for different years).
   - **Example**: Comparing sales for multiple products over time.

````python
#âœ… Multiple Bar Charts (Grouped Bar Chart)
import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 45, 56, 78, 33]
values2 = [14, 36, 44, 60, 29]

# Set bar width and positions
bar_width = 0.35
index = np.arange(len(categories))

# Create grouped bar chart
plt.bar(index, values1, bar_width, label='Group 1', color='blue')
plt.bar(index + bar_width, values2, bar_width, label='Group 2', color='red')

# Add title and labels
plt.title("Grouped Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.xticks(index + bar_width / 2, categories)
plt.legend()

# Show the plot
plt.show()
````

````python
#âœ… Customized Grouped Bar Chart Example

import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 45, 56, 78, 33]
values2 = [14, 36, 44, 60, 29]

# Set bar width and positions
bar_width = 0.35
index = np.arange(len(categories))

# Create grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))  # Customize the figure size

# Plot the bars
bar1 = ax.bar(index, values1, bar_width, label='Group 1', color='steelblue')
bar2 = ax.bar(index + bar_width, values2, bar_width, label='Group 2', color='lightcoral')

# Add value labels on top of bars
for bar in bar1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 1, str(yval), ha='center', va='bottom', fontsize=12)

for bar in bar2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 1, str(yval), ha='center', va='bottom', fontsize=12)

# Add title and labels
ax.set_title("Customized Grouped Bar Chart", fontsize=14, fontweight='bold')
ax.set_xlabel("Category", fontsize=12)
ax.set_ylabel("Values", fontsize=12)

# Customize ticks
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(categories, fontsize=12)

# Show gridlines for better readability
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

# Add legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
````

####  **ðŸ”¹Stacked Bar Chart**
   - **Description**: Displays multiple data sets stacked on top of each other within the same bar.
   - **Use Case**: Ideal when showing parts of a whole (e.g., sales composition over time).
   - **Example**: Showing sales for different products across multiple years.

````python
# Stacked Bar Chart

import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 45, 56, 78, 33]
values2 = [14, 36, 44, 60, 29]

# Set bar width and positions
bar_width = 0.5
index = np.arange(len(categories))

# Create stacked bar chart
plt.bar(index, values1, bar_width, label='Group 1', color='skyblue')
plt.bar(index, values2, bar_width, bottom=values1, label='Group 2', color='coral')

# Add title and labels
plt.title("Stacked Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.xticks(index, categories)
plt.legend()

# Show the plot
plt.show()
````

####  **ðŸ”¹100% Stacked Bar Chart**
   - **Description**: Similar to a stacked bar chart, but the heights of all bars add up to 100%, making it easier to see relative percentages.

   - **Use Case**: Useful for comparing the proportion of categories across multiple groups.
   - **Example**: Comparing market share across different companies.

````python
# 100% Stacked Bar Chart

import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 45, 56, 78, 33]
values2 = [14, 36, 44, 60, 29]

# Calculate the total values for each category
total_values = np.array(values1) + np.array(values2)

# Calculate the percentages for each group
percentage1 = (np.array(values1) / total_values) * 100
percentage2 = (np.array(values2) / total_values) * 100


# Set bar width and positions
bar_width = 0.5
index = np.arange(len(categories))

# Create 100% stacked bar chart
plt.bar(index, percentage1, bar_width, label='Group 1', color='skyblue')
plt.bar(index, percentage2, bar_width, bottom=percentage1, label='Group 2', color='coral')

# Add title and labels
plt.title("100% Stacked Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Percentage (%)")
plt.xticks(index, categories)
plt.legend()

# Show the plot
plt.show()
````

## ðŸ¥§ Pie Charts in Matplotlib

#### ðŸ”¹ What is a Pie Chart?
A **Pie Chart** is a circular statistical diagram, divided into slices to illustrate **numerical proportion**. Each slice represents a categoryâ€™s contribution to the whole.

---

#### ðŸ”¹ When to Use a Pie Chart

| âœ… Best Used For                        | âŒ Avoid When                         |
|----------------------------------------|--------------------------------------|
| Showing part-to-whole relationships    | Comparing many categories            |
| Visualizing percentage distributions   | When data includes negative values   |
| Displaying simple categorical data     | When precise comparison is required  |


# **Basic Pie Chart**

````python
import matplotlib.pyplot as plt

# Data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]

# Pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Basic Pie Chart")
plt.show()
````

#### ðŸ”¹ Types & Variants of Pie Charts

---

**1. Standard Pie Chart**  
- **Use Case**: Simple percentage or portion breakdowns.  
- **Code**: See above.

---

**2. Exploded Pie Chart**  
- **Description**: "Pulls out" one or more slices to emphasize them.  
- **Use Case**: Highlight the most important slice.


````python
# "Explode" first slice (Apples)
explode = (0.1, 0, 0, 0)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)
plt.title("Exploded Pie Chart")
plt.show()
````

**3. Donut Chart (Pie with a Hole)**  
- **Description**: Pie chart with a circular hole in the middle.  
- **Use Case**: Aesthetic alternative to pie chart.


````python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
# Add a white circle to make it a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Donut Chart")
plt.show()
````

# **Pie Chart with Custom Colors and Shadow**

````python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Customized Pie Chart")
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
````

---

| `colors`      | Custom color list for slices                     |

## ðŸ“Š Histograms in Matplotlib

#### ðŸ”¹ What is a Histogram?

A **Histogram** is a graphical representation of the distribution of numerical data. It groups the data into **bins** (intervals) and shows how many values fall into each bin using bars.

---

#### ðŸ”¹ When to Use a Histogram

| âœ… Best Used For                       | âŒ Avoid When                        |
|---------------------------------------|-------------------------------------|
| Understanding data distribution       | Displaying categorical data         |
| Spotting skewness or outliers         | Needing exact value comparisons     |
| Comparing frequency across intervals  | Data is not numeric or continuous   |


# **Basic Histogram Example**

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
data = np.random.randn(1000)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

---

## ðŸ”¹ Types & Variants of Histograms in Matplotlib

---

### 1. **Standard Histogram**
- **Description**: The most basic form; shows frequency distribution using bars.
- **Use Case**: Understand distribution shape, range, and frequency.

---

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='coral', edgecolor='black')
plt.title('Standard Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

### 2. **Stacked Histogram**
- **Description**: Multiple datasets stacked on top of each other.
- **Use Case**: Comparing multiple distributions in one chart.

---


````python
# Generate two datasets
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2

plt.hist([data1, data2], bins=30, stacked=True, label=['Data1', 'Data2'], color=['skyblue', 'orange'], edgecolor='black')
plt.title('Stacked Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
````

### 3. **Normalized Histogram (Density Plot)**
- **Description**: Area under histogram sums to 1 instead of showing frequency.
- **Use Case**: Comparing datasets of different sizes; probability distribution.

---

````python
plt.hist(data, bins=30, density=True, color='green', edgecolor='black')
plt.title('Normalized Histogram')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
````

### 4. **Horizontal Histogram**
- **Description**: Bars are drawn horizontally instead of vertically.
- **Use Case**: Better layout for long x-axis labels or horizontal comparison.

---

````python
plt.hist(data, bins=30, orientation='horizontal', color='purple', edgecolor='black')
plt.title('Horizontal Histogram')
plt.xlabel('Frequency')
plt.ylabel('Value')
plt.grid(True)
plt.show()
````

### 5. **Custom Bins**
- **Description**: Manually defined bin ranges.
- **Use Case**: Precise control over how data is grouped.

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='coral', edgecolor='black')
plt.title('Standard Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

---

## âœ³ï¸ Scatter Plots in Matplotlib

#### ðŸ”¹ What is a Scatter Plot?

A **Scatter Plot** displays values for two numeric variables as a collection of points. Each point represents an observation in the dataset, plotted along the x and y axes.

---

#### ðŸ”¹ When to Use a Scatter Plot

| âœ… Best Used For                             | âŒ Avoid When                         |
|---------------------------------------------|--------------------------------------|
| Visualizing relationships between variables | When variables are categorical       |
| Detecting trends, patterns, or clusters     | Plotting large dense datasets        |
| Identifying correlation or outliers         | Comparing summary statistics         |


# **Basic Scatter Plot Example**

---

## ðŸ”¹ Types & Variants of Scatter Plots in Matplotlib

---

### 1. **Basic Scatter Plot**
- **Description**: Plots x vs y as individual points.
- **Use Case**: Simple 2D relationship between variables.

---

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.random.rand(100)
y = x * 2 + np.random.normal(0, 0.1, 100)

# Basic scatter plot
plt.scatter(x, y, color='dodgerblue')
plt.title('Basic Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
````

### 2. **Colored Scatter Plot**
- **Description**: Points are colored based on a third variable or category.
- **Use Case**: Show categories or magnitude through color.

---


````python
colors = np.random.choice(['red', 'green', 'blue'], 100)

plt.scatter(x, y, c=colors)
plt.title('Colored Scatter Plot by Category')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
````

### 3. **Sized Scatter Plot**
- **Description**: Point sizes reflect a third numerical value.
- **Use Case**: Highlight magnitude or weight visually.

---

````python
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, s=sizes, color='orange', alpha=0.6, edgecolors='black')
plt.title('Sized Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
````

### 4. **Labeled Scatter Plot**
- **Description**: Each point has a text label for identification.
- **Use Case**: Annotating important points or IDs.

---

````python
plt.scatter(x, y, color='teal')

for i in range(10):  # Label only first 10 points
    plt.text(x[i], y[i], f'P{i}', fontsize=9)

plt.title('Labeled Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
````

### 5. **Scatter Plot with Colormap**
- **Description**: Uses a gradient colormap to encode a third variable.
- **Use Case**: Visualize intensity, temperature, etc.

````python
z = np.random.randn(100)

plt.scatter(x, y, c=z, cmap='viridis')
plt.title('Scatter Plot with Colormap')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.colorbar(label='Value')
plt.grid(True)
plt.show()
````

---

## ðŸ”¹ Scatter Plot Parameters Overview

| Parameter       | Description                                                |
|-----------------|------------------------------------------------------------|
| `x`             | Data for the x-axis                                        |
| `y`             | Data for the y-axis                                        |
| `c`             | Color of the points (can be a sequence or a single color)   |
| `s`             | Size of the points                                         |
| `cmap`          | Colormap for the `c` values (e.g., 'viridis', 'plasma')    |
| `alpha`         | Transparency level of the points (0.0 to 1.0)              |
| `edgecolors`    | Color of the point edges                                    |
| `linewidths`    | Width of the point edges                                    |
| `marker`        | Shape of the points (e.g., 'o', '^', 's', etc.)             |
| `label`         | Legend label for the scatter plot                           |

---

## ðŸ”¹ Summary

| Scatter Plot Type         | Use When...                                         |
|---------------------------|-----------------------------------------------------|

| Sized Scatter Plot        | Visualizing the impact or weight of a variable through point size |
| Labeled Scatter Plot      | Annotating points with labels for easy identification |

### ðŸ”¥ 6. Seaborn Heatmap

A **heatmap** is a graphical representation of 2D data where individual values are represented by varying **color intensities**. It is widely used to **visualize correlation matrices**, feature relationships, or any large grid-based data.

---

#### ðŸ”¹ When to Use:
- âœ… Visualizing correlation matrices  
- âœ… Showing patterns in large datasets  

---

#### ðŸ”¸ Example:

```python



````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flights = sns.load_dataset("flights")

# Create a pivot table using keyword arguments
pivot_data = flights.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
sns.heatmap(pivot_data, annot=True, fmt="d", cmap="YlGnBu")

# Customize plot
plt.title("Seaborn Heatmap - Flight Passengers")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
````

### ðŸ“ˆ 7. Seaborn KDE Plot (Kernel Density Estimate)

A **KDE plot** is a smoothed, continuous version of a histogram. It estimates the **probability density function** of a continuous variable and is useful for visualizing the underlying distribution.

---

#### ðŸ”¹ When to Use:
- âœ… Estimating the distribution of a continuous variable  
- âœ… Visualizing the underlying probability distribution of the data  

---

#### ðŸ”¸ Example:

```python


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a KDE plot
sns.kdeplot(data=tips["total_bill"], fill=True, color="skyblue")

# Customize the plot
plt.title('Seaborn KDE Plot')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
````

### ðŸ”¢ 8. Seaborn Count Plot

A **count plot** is a bar plot that shows the **frequency of each category** in a **categorical variable**. Unlike a typical bar plot that summarizes numerical values, a count plot counts the **occurrences** of each category.

#### ðŸ”¹ When to Use:

### ðŸŽ¯ 9. Seaborn Strip Plot

A **strip plot** shows **individual data points** along a **categorical axis**. It's ideal for **small datasets** where displaying every point is feasible. It helps visualize the **distribution** and **spread** within each category.

---

#### ðŸ”¹ When to Use:
- âœ… Visualizing the distribution of small datasets  
- âœ… Displaying individual data points in a category  

---

#### ðŸ”¸ Example:

```python



````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a strip plot
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, palette="Set2")

# Customize the plot
plt.title('Seaborn Strip Plot')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
````

### ðŸ 10. Seaborn Swarm Plot

A **swarm plot** is similar to a strip plot, but it **automatically adjusts point positions** to avoid overlap. This makes it ideal for **small datasets** where itâ€™s important to show **all individual data points clearly**.

---

#### ðŸ”¹ When to Use:
- âœ… Visualizing the distribution of small datasets while avoiding overlapping points  
- âœ… Displaying all individual data points without clustering them  

---

#### ðŸ”¸ Example:

```python


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a swarm plot
sns.swarmplot(x="day", y="total_bill", data=tips, palette="muted")

# Customize the plot
plt.title('Seaborn Swarm Plot')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
````
